## Activity 1: Sentiment Analysis with a Pretrained Transformer

Remember how in Week 25 you had to:
1. Tokenize the text
2. Remove stopwords
3. Extract TF-IDF features
4. Train a Logistic Regression model
5. Then predict

With a pretrained Transformer, **all of that happens automatically in one line!**

```python
# =============================================
# SENTIMENT ANALYSIS — One line of code!
# =============================================

from transformers import pipeline

# Load a pretrained sentiment analysis model
# (This downloads the model the first time — may take a minute)
print("Loading pretrained sentiment model...")
classifier = pipeline("sentiment-analysis")
print("Model loaded!\n")

# Test with sample reviews
reviews = [
    "This movie was absolutely incredible! A true masterpiece.",
    "Worst film I've ever seen. Complete waste of time and money.",
    "It was okay. Some good moments but overall pretty average.",
    "The acting was phenomenal and the storyline kept me on the edge of my seat!",
    "Boring, predictable, and way too long. I almost fell asleep.",
]

print("Sentiment Analysis Results:")
print("=" * 65)
for review in reviews:
    result = classifier(review)[0]
    emoji = "👍" if result['label'] == 'POSITIVE' else "👎"
    print(f'{emoji} {result["label"]} ({result["score"]:.1%}): "{review[:55]}..."')
print("=" * 65)
```

```python
# =============================================
# TRY YOUR OWN SENTENCES!
# =============================================
# Change these to anything you want — movie reviews,
# restaurant reviews, comments about school, etc.

my_sentences = [
    "I really enjoyed learning about neural networks this year!",
    "Homework is the worst thing ever invented.",
    "The pizza was decent but the service was slow.",
]

print("Your sentences:")
print("-" * 50)
for sentence in my_sentences:
    result = classifier(sentence)[0]
    emoji = "👍" if result['label'] == 'POSITIVE' else "👎"
    print(f'{emoji} {result["label"]} ({result["score"]:.1%})')
    print(f'   "{sentence}"')
    print()
```

## Activity 2: Text Generation with GPT-2

Now let's see a Transformer **generate text**! We'll use GPT-2 (a smaller version of ChatGPT's ancestor). Give it a prompt and watch it continue writing.

```python
# =============================================
# TEXT GENERATION — Watch a Transformer write!
# =============================================

from transformers import pipeline

print("Loading GPT-2 text generation model...")
generator = pipeline("text-generation", model="gpt2")
print("Model loaded!\n")

# Try different prompts!
prompt = "The future of artificial intelligence is"

print(f'Prompt: "{prompt}"')
print("=" * 60)

outputs = generator(
    prompt,
    max_length=60,           # Maximum total length (prompt + generated)
    num_return_sequences=3,  # Generate 3 different completions
    do_sample=True,          # Enable randomness for variety
    temperature=0.8,         # Controls randomness (lower = more predictable)
)

for i, output in enumerate(outputs):
    print(f"\nGeneration {i+1}:")
    print(f"  {output['generated_text']}")

print("\n" + "=" * 60)
print("Notice: Each generation is different because the model samples")
print("from probabilities — this is the 'temperature' parameter at work!")
```

```python
# =============================================
# TRY YOUR OWN PROMPTS!
# =============================================
# Change the prompt and see what the model generates.
# Try different topics, styles, and sentence starters.

my_prompts = [
    "Once upon a time in a land far away",
    "The best way to learn machine learning is",
    "In the year 2050, schools will",
]

for prompt in my_prompts:
    output = generator(prompt, max_length=50, num_return_sequences=1, do_sample=True, temperature=0.7)
    print(f'Prompt: "{prompt}"')
    print(f'  -> {output[0]["generated_text"]}')
    print()
```

```python
# =============================================
# EXPERIMENT: How Temperature Affects Output
# =============================================
# Temperature controls randomness:
#   Low (0.2) = very predictable, repetitive
#   Medium (0.7) = balanced
#   High (1.5) = very creative, sometimes nonsensical

prompt = "Machine learning is"

print(f'Prompt: "{prompt}"\n')

for temp in [0.2, 0.7, 1.5]:
    output = generator(prompt, max_length=40, num_return_sequences=1,
                       do_sample=True, temperature=temp)
    text = output[0]['generated_text']
    label = "LOW (predictable)" if temp < 0.5 else "MEDIUM (balanced)" if temp < 1.0 else "HIGH (creative)"
    print(f"Temperature {temp} — {label}:")
    print(f"  {text}")
    print()
```

## Activity 3: Explore More Transformer Tasks

Transformers can do much more than sentiment analysis and text generation. Let's try summarization, named entity recognition, and question answering.

```python
# =============================================
# SUMMARIZATION — Condense long text automatically
# =============================================

from transformers import pipeline

print("Loading summarization model...")
summarizer = pipeline("summarization")
print("Model loaded!\n")

long_text = """
Artificial intelligence has made remarkable progress in recent years, 
particularly in the field of natural language processing. Large language 
models like GPT-4 and Claude can now write essays, answer complex questions, 
generate computer code, and even engage in sophisticated reasoning tasks. 
These models are based on the Transformer architecture, which was first 
introduced in a 2017 paper titled "Attention Is All You Need" by researchers 
at Google. The key innovation of the Transformer is the self-attention 
mechanism, which allows the model to consider all parts of the input text 
simultaneously rather than processing it sequentially. This parallel 
processing makes Transformers much faster to train than previous approaches 
and enables them to capture long-range dependencies in text more effectively.
"""

print(f"Original text: {len(long_text.split())} words")
print(f"  {long_text.strip()[:200]}...\n")

summary = summarizer(long_text, max_length=60, min_length=20, do_sample=False)
print(f"Summary: {len(summary[0]['summary_text'].split())} words")
print(f"  {summary[0]['summary_text']}")
```

```python
# =============================================
# NAMED ENTITY RECOGNITION (NER)
# =============================================
# Find and classify names, places, and organizations in text

from transformers import pipeline

print("Loading NER model...")
ner = pipeline("ner", grouped_entities=True)
print("Model loaded!\n")

sentences = [
    "Elon Musk founded SpaceX in Hawthorne, California in 2002.",
    "The Eiffel Tower in Paris, France attracts millions of visitors each year.",
    "Google and Microsoft are competing to build the best AI systems.",
]

for sentence in sentences:
    print(f'Text: "{sentence}"')
    results = ner(sentence)
    for entity in results:
        print(f'  -> "{entity["word"]}" = {entity["entity_group"]} ({entity["score"]:.1%})')
    print()
```

```python
# =============================================
# QUESTION ANSWERING
# =============================================
# Give the model a context paragraph and ask questions about it

from transformers import pipeline

print("Loading question-answering model...")
qa = pipeline("question-answering")
print("Model loaded!\n")

context = """
The Transformer architecture was introduced in 2017 by researchers at Google 
in a paper called "Attention Is All You Need." The key innovation is the 
self-attention mechanism, which allows every word to look at every other word 
in the input. ChatGPT, made by OpenAI, and Claude, made by Anthropic, are 
both based on the Transformer architecture. These models are called Large 
Language Models (LLMs) because they have billions of parameters and are 
trained on massive amounts of text data.
"""

questions = [
    "When was the Transformer architecture introduced?",
    "What is the key innovation of the Transformer?",
    "Who made Claude?",
    "What are LLMs trained on?",
]

print("Context paragraph loaded. Asking questions...\n")
for q in questions:
    result = qa(question=q, context=context)
    print(f'Q: {q}')
    print(f'A: "{result["answer"]}" (confidence: {result["score"]:.1%})')
    print()
```

## Activity 4: The Ultimate Comparison — Classic ML vs. Transformer

Let's put it all together by comparing the Week 25 approach (TF-IDF + Logistic Regression) with a pretrained Transformer on the exact same sentiment analysis task.

```python
# =============================================
# CLASSIC ML vs PRETRAINED TRANSFORMER
# =============================================
# Same task. Two eras of AI. Head to head!

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from transformers import pipeline
import time

# Test reviews — deliberately tricky!
test_reviews = [
    "This movie was absolutely fantastic! A true masterpiece of cinema.",
    "Terrible waste of time. The acting was wooden and the plot was nonsensical.",
    "Not bad, but not great either. Some decent moments scattered throughout.",
    "I didn't expect to like it, but it completely won me over. Brilliant!",
    "The trailer looked promising, but the movie itself was a huge letdown.",
]

# ---- APPROACH 1: Classic ML ----
print("=" * 65)
print("APPROACH 1: Classic ML (TF-IDF + Logistic Regression)")
print("  - Requires manual feature engineering")
print("  - Must train on labeled data")
print("=" * 65)

# Simple training data
train_texts = [
    "great movie loved it fantastic wonderful amazing brilliant",
    "terrible awful bad horrible waste boring worst dull",
    "good film nice enjoyable great fun entertaining superb",
    "bad movie terrible acting poor writing dull lifeless",
    "excellent masterpiece outstanding incredible beautiful",
    "worst garbage trash disappointing painful unwatchable",
] * 20
train_labels = [1, 0, 1, 0, 1, 0] * 20

tfidf = TfidfVectorizer(stop_words='english')
X_train = tfidf.fit_transform(train_texts)
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, train_labels)

X_test = tfidf.transform(test_reviews)
lr_preds = lr_model.predict(X_test)

print()
for review, pred in zip(test_reviews, lr_preds):
    label = "POSITIVE" if pred == 1 else "NEGATIVE"
    emoji = "👍" if pred == 1 else "👎"
    print(f'  {emoji} {label}: "{review[:55]}..."')

# ---- APPROACH 2: Pretrained Transformer ----
print(f'\n{"=" * 65}')
print("APPROACH 2: Pretrained Transformer")
print("  - No manual feature engineering needed")
print("  - No training required — uses transfer learning!")
print("=" * 65)

classifier = pipeline("sentiment-analysis")
tf_results = classifier(test_reviews)

print()
for review, result in zip(test_reviews, tf_results):
    emoji = "👍" if result['label'] == 'POSITIVE' else "👎"
    print(f'  {emoji} {result["label"]} ({result["score"]:.1%}): "{review[:50]}..."')

# ---- SCOREBOARD ----
print(f'\n{"=" * 65}')
print("SCOREBOARD")
print("=" * 65)
print(f"{'Feature':<30} {'Classic ML':<20} {'Transformer'}")
print("-" * 65)
print(f"{'Feature engineering':<30} {'Manual (TF-IDF)':<20} {'Automatic'}")
print(f"{'Training required':<30} {'Yes':<20} {'No (pretrained)'}")
print(f"{'Lines of code':<30} {'~15':<20} {'~3'}")
print(f"{'Handles tricky reviews':<30} {'Struggles':<20} {'Much better'}")
print(f"{'Confidence scores':<30} {'Basic':<20} {'Detailed'}")
print("-" * 65)
print()
print("The Transformer wins on accuracy, especially on tricky sentences.")
print("But Classic ML is still great for learning the fundamentals —")
print("which is exactly what we did in Weeks 1-25!")
```
